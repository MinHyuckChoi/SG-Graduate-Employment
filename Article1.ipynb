# Which University in Singapore lives up to its name?

In December 2007, a legendary investor, Jim Rogers, sold his mansion in New York City and decided to move to Singapore due to his two daughters’ education. According to the QS Asia University Ranking 2018, Nanyang Technological University is ranked the best university in Asia and it is followed by National University of Singapore. Not surprisingly, the competition in education is not limited to Singaporean children and tremendous money is spent on education. Literally, it is one of the most education-oriented countries in the world.

Now, one question comes to our mind: Which university in Singapore lives up to its name? To see if their time and cost are really worth, it is our utmost interest to check the employment rate and salary of graduates. Fortunately, the Singapore government opens related data to the public.

Before starting the analysis, the values I used for this analysis needs to be clear. I chose Full-Time Permanent Employment Rate (%) as employment index, and Gross Monthly Salary Mean (S$) as salary index. In addition, missing values in the data set are eliminated for accurate analysis.

To shape the outline, the two charts below show that the graduate’s employment rate of full-time permanent dropped the last two years, while the monthly gross amount steadily increases.

Our point is to identify the best university among those five universities; National University of Singapore(NUS), Singapore Institute of Technology(SIT), Nanyang Technological University(NTU), Singapore Management University(SMU) and Singapore University of Technology and Design(SUTD). Plotting by each university can give us visual insight. Although all institutions represent a high employment rate, SMU records the highest score of both categories, employment rate and gross salary.

A closer look is in need before jumping to a conclusion. Plotting by year, these plots reveal the time series tendency by universities in a glance. While the gross salary of each university has increased, employment rate part looks a bit complicating. However, SUTD has notably increased 4th to the highest out of five universities for two years.



Next, for the exploratory data analysis, below are the EDA boxplots. In boxplot of gross salary by universities, we can see that SMU has some extraordinary departments, which earns over SGD 5,000 per month. Ordering by gross salary, the table shows that school of law – especially SMU – extraordinary outruns the other schools, and it guarantees the highest employment rate and gross salary after graduation.

Lastly, plotting this data on a map can represent us if there is spatial relation. It shows that all five universities are located in the southern part of Singapore, and the low gross salary of SIT relatively stands out in the rest.

Now, it is clear that SMU is ranked as the highest employment rate and monthly gross salary. However, through data visualization, we also discovered two important things. One is that school of law – especially SMU - is ranked as the highest gross salary out of the five universities. The other is that the recent growth of SUTD is remarkable in both employment rate and gross salary.
